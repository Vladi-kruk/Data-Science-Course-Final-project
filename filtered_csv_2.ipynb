```python
df_final= pd.DataFrame()
df2=df.copy()
cities = set(df2['city'])
years = set(df2['year'])
rooms = set(df2['rooms'])
kinds = set(df2['kind'])

for year in years:
    for city in cities:
        for kind in kinds:
            for room in rooms:
                df_test = df2[df2['year']==year]
                df_test = df_test[df_test['city']==city]
                df_test = df_test[df_test['kind']==kind]
                df_test = df_test[df_test['rooms']==room]
                if len(df_test['price'])==0:
                    continue
                Q1 = np.percentile(df_test['price'], 25)
                Q3 = np.percentile(df_test['price'], 75)
                IQR = Q3-Q1
                outliers_rows = df_test[(df_test['price']<Q1 - 1.5*IQR) | (df_test['price'] > Q3 + 1.5*IQR)].index
                df_test.drop(labels = outliers_rows , inplace=True)
                df_final = df_final.append(df_test, ignore_index=True)
    
df_final
```

```python
df_final.dropna(axis='rows', inplace=True)
```
